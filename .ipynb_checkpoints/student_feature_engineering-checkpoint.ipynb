```python
# making new columns for better understanding-- 
df["total_score"]=df["math score"]+df['reading score']+df["writing score"] #for total marks
df['average_score']=df["total_score"]/3    # for average marks
```

```python
df.head(5)
```

```python
passmarks = 35

df["Percentage"]=(df["math score"]+df['reading score']+df["writing score"])/300*100   #percentage columns 
```

```python
def Grade(Percentage):
    if (Percentage >= 95): return "A++"     # O
    if (Percentage >= 81): return "A+"      # A
    if (Percentage >= 71): return "A"       # B
    if (Percentage >= 61): return "B"       # C
    if (Percentage >= 51): return "C"       # D
    if (Percentage >= 41): return "D"       # E
    else: return "F"                        # F
    

df['grade']=df.apply(lambda x : Grade(x["Percentage"]),axis=1)
```

```python
df.head(2)
```

```python
# checking the no. of student who has scored full marks or the 100 marks in the reading subject--
reading_score=df[df['reading score']==100]["average_score"].count()
reading_score
```

```python
# checking the no. of student who has Scored full marks or the 100 marks in the math subject--
math_score=df[df['math score']==100]["average_score"].count()
math_score
```

```python
# checking the no. of student who has Scored full marks or the 100 marks writing subject--
writing_score=df[df['writing score']==100]["average_score"].count()
writing_score
```

## Checking for Normal Distribution using Transformation.

 Q-Q plot (to check if data is normally distributed or not)

```python
# density plot using seaborn liabary.
fig, axs= plt.subplots(2,2, figsize=(15,10))

sns.kdeplot(df['math score'],shade=True,label="Math",color="gold",ax=axs[0,0],alpha=0.5)
sns.kdeplot(df['reading score'],shade=True,label="Reading",color='indigo',ax=axs[0,1],alpha=0.5)
sns.kdeplot(df["writing score"],shade=True,label="Writing",color="pink",ax=axs[1,0],alpha=0.5)
sns.kdeplot(df["average_score"],shade=True,label="Average score",color="slateblue",ax=axs[1,1],alpha=0.5)
plt.show()
```

observation:
    
    Here we can observe that all the data are Normal Distributed.

## Box-Cox Transformation

```python
import scipy.stats as stats
```

```python
def qq_plot(column):
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(df[column])
    plt.title("{} PDF".format(column)) #Probablity density function
    plt.subplot(122)
    stats.probplot(df[column],dist="norm",plot=plt)
    plt.title("{} QQ Plot".format(column))
    plt.show()
```

```python
qq_plot("writing score")
```

```python
qq_plot("reading score")
```

```python
qq_plot("math score")
```

```python
qq_plot("average_score")
```

Observation:
    
    For range -2 to 2 average_score and math_score follows normal distribution, but for 
    value less than -2 and for values greater than 2 it doesn't follow normal distribution.
